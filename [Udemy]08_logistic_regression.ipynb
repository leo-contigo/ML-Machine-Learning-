## Importing the dataset

## 데이터 불러옴

```
dataset = pd.read_csv('파일위치/Part 3 - Classification(분류)/Section 14 - Logistic Regression/Social_Network_Ads.csv')
dataset
```

## Splitting the dataset into the Training set and Test set
---
## 훈련, 테스트 세트 나누기

```
from sklearn.model_selection import train_test_split

# 400개 데이터에서 25%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```
print(len(X_train))
print("="*50)
print(X_train)
```

```
print(len(y_train))
```

```
print(X_test)
print("="*50)
print(len(X_test))
```

```
print(y_test)
print("="*50)
print(len(y_test))
```

## Feature Scaling
---
## 스케일링

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

# 훈련
X_train = sc.fit_transform(X_train)

# 테스트
X_test = sc.transform(X_test)
```

```
print(X_train)
```

```
print(X_test)
```

## Training the Logistic Regression model on the Training set
---


```
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state = 0)

# 훈련
classifier.fit(X_train, y_train)
```

## Predicting a new result
---
단일 결과의 구매결정 예측<br>
- Q) 30세, 예상급여 $87,000
- 위 사람의 구매 여부? 예측 > 예측결과와 실제 결과 비교


```
# 예측
print(classifier.predict(sc.transform( [[30,87000]] )))
```

## Predicting the Test set results
---
테스트 결과 예측

```
y_pred = classifier.predict(X_test)

print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))

# [예측 결과 실제 구매 결정]
```

## Making the Confusion Matrix
---
[sklearn metrics]('https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html#sklearn.metrics.confusion_matrix')

```
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
print(cm)

# 정확도
accuracy_score(y_test, y_pred)

# 테스트 세트의 고객에 대해 65개 올바른 예측, 잘못된 예측 3건(구매하지 않음 예측 > 구매)
# 잘못된 예측 8건(구매하지 않음 > 구매한다고 예측)
# 정확도 89%
```

## Visualising the Training set results
---
시각화

```
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show();
```

## Visualising the Test set results

```
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show();
```
